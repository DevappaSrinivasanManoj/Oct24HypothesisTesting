```python
from scipy.stats import ttest_rel

# Given data
before = [2.45, 0.69, 1.80, 2.80, 0.07, 1.67, 2.93, 0.47, 1.45, 1.34]
after = [7.71, 2.17, 5.65, 8.79, 0.23, 5.23, 9.19, 1.49, 4.56, 4.20]

t_stat, p_value_one_tailed = ttest_rel(before, after, alternative='less')

# Output the t-statistic and one-tailed p-value
print(f"T-statistic: {t_stat:.4f}")
print(f"One-tailed p-value: {p_value_one_tailed:.4f}")
```

```python
from scipy.stats import ttest_rel

# Given data
before = [2.45, 0.69, 1.80, 2.80, 0.07, 1.67, 2.93, 0.47, 1.45, 1.34]
after = [7.71, 2.17, 5.65, 8.79, 0.23, 5.23, 9.19, 1.49, 4.56, 4.20]

# Performing a one-tailed paired t-test (alternative='greater')
t_stat, p_value_one_tailed = ttest_rel(after, before, alternative='greater')

# Output the t-statistic and one-tailed p-value
print(f"T-statistic: {t_stat:.4f}")
print(f"One-tailed p-value: {p_value_one_tailed:.4f}")
```
